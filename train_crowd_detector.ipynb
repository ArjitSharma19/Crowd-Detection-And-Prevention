# Local YOLOv8 Crowd Detector Training Notebook

This notebook runs the training process for your Crowd Detection model locally on your laptop's NVIDIA GeForce RTX 4060 GPU.

### Steps:
1. Verify PyTorch CUDA GPU availability
2. Train the YOLOv8 model on your local dataset

## Step 1: Verify GPU Availability
Run this cell to confirm that Python is using your dedicated NVIDIA graphics card.

```python
import torch
import ultralytics
ultralytics.checks()
```

## Step 2: Train YOLOv8 Model

We will start with the pre-trained weights (`yolov8n.pt` - YOLOv8 Nano, which is fast and lightweight) and train it on your local dataset using your RTX 4060.

- `data`: Path to your dataset's `data.yaml` config file
- `epochs`: Number of passes over the dataset (50 is standard for fine-tuning)
- `imgsz`: Image resolution (640 is standard)
- `device`: Device to run on (device=0 selects the RTX 4060 GPU)
- `batch`: Batch size (set to 4 to prevent OutOfMemory issues on laptops)
- `workers`: Number of data loading worker threads (set to 0 on Windows to prevent dataloader deadlocks/freezes)

```python
from ultralytics import YOLO

# Load a pre-trained YOLOv8 Nano model
model = YOLO('yolov8n.pt')

# Train the model
results = model.train(
    data='E:/Crowd Detection &  Prevention/crowd density.v1-v1.yolov8/data.yaml',
    epochs=50,
    imgsz=640,
    device=0,
    batch=4,             # Safe batch size for laptops
    workers=0,           # Set to 0 on Windows to prevent multiprocessing freezes
    project='runs/detect' # Saves training runs locally inside the project folder
)
```
