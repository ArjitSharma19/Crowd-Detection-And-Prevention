# Local YOLOv8 Crowd Detector Training Notebook

This notebook runs the training process for your Crowd Detection model locally on your laptop's NVIDIA GeForce RTX 4060 GPU.

### Steps:
1. Verify PyTorch CUDA GPU availability
2. Train the YOLOv8 model on your local dataset

## Step 1: Verify GPU Availability
Run this cell to confirm that Python is using your dedicated NVIDIA graphics card.

````python
import torch
import ultralytics
ultralytics.checks()
````

## Step 2: Train YOLO Model (Notebook-based)

We will start with the pre-trained weights for an industry-standard model scale like **YOLO11 Medium (`yolo11m.pt`)** or **YOLOv8 Medium (`yolov8m.pt`)** and fine-tune it on your dataset.

**Key Hyperparameters:**
- `model`: `'yolo11m.pt'` (latest, more accurate) or `'yolov8m.pt'`
- `data`: Path to your dataset's `data.yaml` config file
- `epochs`: Number of passes over the dataset (100 is recommended for industry standards)
- `batch`: Batch size (set to 8 or 4 depending on VRAM usage)
- `device`: Device to run on (device=0 selects the RTX 4060 GPU)
- `workers`: Number of data loading worker threads (set to 0 on Windows to prevent deadlocks)

````python
from ultralytics import YOLO

# Load an industry-standard YOLO11 Medium model (pretrained on COCO)
model = YOLO('yolo11m.pt')

# Train the model locally on your RTX 4060 GPU
results = model.train(
    data='E:/Crowd Detection &  Prevention/crowd density.v1-v1.yolov8/data.yaml',
    epochs=100,
    imgsz=640,
    device=0,
    batch=8,              # Set to 8 for medium scale, reduce to 4 if OOM occurs
    workers=0,            # Set to 0 on Windows to prevent multiprocessing deadlocks
    project='runs/detect',
    name='train_yolo11m'
)
````
